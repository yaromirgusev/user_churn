# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from catboost import CatBoostClassifier, Pool, cv
```

## Пункт 1

```python
df_new = pd.read_csv('./train.csv')
both1=[]
for i in df_new:
    if i!='Churn':
        both1.append(i)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_new[both1], df_new['Churn'], random_state=42, train_size=0.8,stratify=df_new['Churn'])
```

## Пункт 2

```python
model = CatBoostClassifier(
    task_type='GPU',
    random_seed=42
)

model.fit(
    x_train, y_train,
    cat_features=cat_cols,
    verbose=False
)
y_prob=model.predict(x_test)
print(roc_auc_score(y_test,y_prob))
```

```python
model1 = CatBoostClassifier(
    iterations=70,
    random_seed=42,
    learning_rate=0.65,
    depth=8,
    l2_leaf_reg=6,
    thread_count=10
)

model1.fit(
    x_train, y_train,
    cat_features=cat_cols,
    verbose=False
)
y_prob=model1.predict(x_test)
print(roc_auc_score(y_test,y_prob))
```

```python
datapool = Pool(data=df_new[both1], label=df_new['Churn'], cat_features=cat_cols)
```

```python
model2=CatBoostClassifier(
    iterations=100,
    loss_function='Logloss',
    #task_type='CPU',
    #thread_count=12
    #devices='0:1'
)
pars={
    'depth': np.arange(2, 10),
    'learning_rate': np.linspace(0.05,1,6),
    #'l2_leaf_reg': np.linspace(3, 7, 6),
    'min_data_in_leaf': np.arange(1, 10, 2)
}
scores = model2.grid_search(pars, datapool, cv=5,verbose=50)
```

```python
scores['params']
```

```python
model2_upd=CatBoostClassifier(
    iterations=300,
    loss_function='Logloss',
    **scores['params']
    #task_type='CPU',
    #thread_count=12
    #devices='0:1'
)
model2_upd.fit(datapool)
```

```python
roc_auc_score(df_new[target_col], model2_upd.predict_proba(datapool)[:,1])
```

Выпишите, какое лучшее качество и с какими параметрами вам удалось получить

'min_data_in_leaf': 1, 'depth': 2, 'learning_rate': 0.24, качество 0.8670949345595934
